```python
targets = (
    28,
    29,
    30,
    41,
    46,
    49,
    85,
    89,
    94,
    97,
)
```

```python
np.random.seed(0)

for kk in range(100):
    dx = 0.05
    x, y = np.mgrid[-3:3:dx, -3:3:dx]

    fig, axs = plt.subplots(5, 5, figsize=(10, 20), dpi=100)

    temp = np.zeros_like(x)

    n_true_mode = np.random.randint(10, 20)
    scales = []
    means = []
    cosv = []
    for i in range(n_true_mode):
        mean = np.random.rand(2) * 6 - 3
        means.append(mean.tolist())

        # cov_scale = np.random.rand() * 0.2 + 0.1
        # cov = cov_scale * (np.random.rand(2, 2) + np.eye(2))

        cov = 0.2 * np.random.rand(2, 2)
        cov = np.dot(cov, cov.T) + 0.5 * np.eye(2)
        print(cov)
        cosv.append(cov.flatten().tolist())

        # print(mean)

        s = stats.multivariate_normal(mean=mean, cov=cov)
        pos = np.dstack((x, y))

        pdf = s.pdf(pos)
        pdf = pdf / pdf.max()

        

        # scale = np.random.rand() * 1 + 0.1
        scale = (np.random.rand() * 0.5 + 0.1) 
        scales.append([scale])

        temp += scale * pdf

    true_temp = temp.copy()
    
    vmax = None
    pred_temps = []
    for j in range(5):
        pred_temp = true_temp.copy()
        for i in range(10):
            mean = np.random.rand(2) * 6 - 3
            cov = 0.2 * (np.random.rand(2, 2) + np.eye(2))
            # cov = 0.05 * (np.random.rand(2, 2) + np.eye(2))

            s = stats.multivariate_normal(mean=mean, cov=cov)
            pos = np.dstack((x, y))

            pdf = s.pdf(pos)    
            pdf = pdf / pdf.max()

            scale = np.random.rand() * 1 + 0.1
            scale /= 1.5 * (j + 1) 

            # sign = np.random.choice([-1, 1])
            sign = 1.0

            pred_temp += sign * scale * pdf

        pred_temps.append(pred_temp)

    max_temp = np.max([true_temp.max(), np.array(pred_temps).max()])
    max_error = np.max(np.abs(true_temp[np.newaxis] - np.array(pred_temps)))
    kwargs = dict(cmap=cmap, vmin=0, vmax=max_temp)

    for j in range(5):
        # axs[j, 0].contourf(x, y, pred_temp, levels=20, **kwargs)
        # axs[j, 1].contourf(x, y, true_temp, levels=20, **kwargs)

        axs[j, 0].pcolormesh(x, y, pred_temps[j], antialiased=True, rasterized=True, **kwargs)
        axs[j, 1].pcolormesh(x, y, true_temp, antialiased=True, rasterized=True, **kwargs)
        # means = np.array(means)
        # axs[j, 2].scatter(means[0], means[1], c='r', s=10)
        
        if vmax is None:
            vmax = np.max(np.abs(true_temp - pred_temp))

        # axs[j, 2].contourf(x, y, np.abs(true_temp - pred_temp), levels=10, cmap=error_cmap, vmin=0, vmax=vmax * 1.2)

        axs[j, 2].pcolormesh(x, y, np.abs(true_temp - pred_temps[j]), antialiased=True, rasterized=True, cmap=error_cmap, vmin=0, vmax=max_error)

        axs[j, 0].set_aspect(1)
        axs[j, 1].set_aspect(1)
        axs[j, 2].set_aspect(1)
        axs[j, 4].set_aspect(1)
        

        axs[j, 0].set_axis_off()
        axs[j, 1].set_axis_off()
        axs[j, 2].set_axis_off()
        axs[j, 3].set_axis_off()
        axs[j, 4].set_axis_off()


    all_data = scales + means + cosv
    axs[-1, 3].text(0, 0, '\n'.join([str([f'{w:.2f}' for w in v]) for v in scales + means + cosv]))

    axs[0, 0].set_title('Predicted')
    axs[0, 1].set_title('True')
    axs[0, 2].set_title('Error')
    axs[0, 4].set_title('Resized')

    # Resize true_temp to 5 x 5 and resize (average value) to original size.
    shape = true_temp.shape
    true_temp_ = cv2.resize(true_temp, (4, 4), interpolation=cv2.INTER_AREA)
    true_temp_ = cv2.resize(true_temp_, (shape[1], shape[0]), interpolation=cv2.INTER_AREA)
    axs[0, 4].pcolormesh(x, y, true_temp_, antialiased=True, rasterized=True, **kwargs)

    true_temp_ = cv2.resize(true_temp, (8, 8), interpolation=cv2.INTER_AREA)
    true_temp_ = cv2.resize(true_temp_, (shape[1], shape[0]), interpolation=cv2.INTER_AREA)
    axs[1, 4].pcolormesh(x, y, true_temp_, antialiased=True, rasterized=True, **kwargs)

    true_temp_ = cv2.resize(true_temp, (16, 16), interpolation=cv2.INTER_AREA)
    true_temp_ = cv2.resize(true_temp_, (shape[1], shape[0]), interpolation=cv2.INTER_AREA)
    axs[2, 4].pcolormesh(x, y, true_temp_, antialiased=True, rasterized=True, **kwargs)

    true_temp_ = cv2.resize(true_temp, (32, 32), interpolation=cv2.INTER_AREA)
    true_temp_ = cv2.resize(true_temp_, (shape[1], shape[0]), interpolation=cv2.INTER_AREA)
    axs[3, 4].pcolormesh(x, y, true_temp_, antialiased=True, rasterized=True, **kwargs)

    if kk in targets:
        plt.savefig(f'./heatmaps/multivariate_normal_{kk}.svg')
        plt.savefig(f'./heatmaps/multivariate_normal_{kk}.png')
        with open(f'./heatmaps/multivariate_normal_{kk}.txt', 'w') as f:
            f.write('\n'.join([str([f'{w:.2f}' for w in v]) for v in all_data]))

        plt.show()
    plt.close()
```

```python
def plot_heatmap(
    seed, n_steps=5, n_true_mode=20, n_error_mode=20, cov=0.3, dx=0.05,
    plot=False, cmap_start=0.08, cmap_end=0.4, ref_cmap='hsv', error_cmap='Blues'
):
    cov_ = cov

    cmap = plt.get_cmap(ref_cmap)
    cmap = mcolors.ListedColormap([cmap(i) for i in np.linspace(cmap_start, cmap_end, 100)][::-1])
    cmap

    error_cmap_ = error_cmap
    error_cmap = plt.get_cmap(error_cmap)
    error_cmap = mcolors.ListedColormap([error_cmap(i) for i in np.linspace(0.01, 1, 100)])

    np.random.seed(seed)

    x, y = np.mgrid[-3:3:dx, -3:3:dx]

    fig, axs = plt.subplots(5, 3, figsize=(10, 20))

    temp = np.zeros_like(x)

    for i in range(n_true_mode):
        mean = np.random.rand(2) * 6 - 3
        cov = 0.3 * (np.random.rand(2, 2) + np.eye(2))

        s = stats.multivariate_normal(mean=mean, cov=cov)
        pos = np.dstack((x, y))

        # scale = np.random.rand() * 1 + 0.1
        scale = 1

        temp += scale * s.pdf(pos)

    true_temp = temp.copy()
    print(true_temp.max())
    kwargs = dict(cmap=cmap, vmin=0, vmax=1)

    vmax = None
    for j in range(n_steps):
        pred_temp = true_temp.copy()
        for i in range(n_error_mode):
            mean = np.random.rand(2) * 6 - 3
            cov = cov * (np.random.rand(2, 2) + np.eye(2))
            # cov = 0.05 * (np.random.rand(2, 2) + np.eye(2))

            s = stats.multivariate_normal(mean=mean, cov=cov)
            pos = np.dstack((x, y))

            scale = np.random.rand() * 1 + 0.1
            scale /= (j + 1)

            # sign = np.random.choice([-1, 1])
            sign = 1.0

            pred_temp += sign *  scale * s.pdf(pos)


        
        # axs[j, 0].contourf(x, y, pred_temp, levels=20, **kwargs)
        # axs[j, 1].contourf(x, y, true_temp, levels=20, **kwargs)

        axs[j, 0].pcolormesh(x, y, pred_temp, antialiased=True, rasterized=True, **kwargs)
        axs[j, 1].pcolormesh(x, y, true_temp, antialiased=True, rasterized=True, **kwargs)
        
        if vmax is None:
            vmax = np.max(np.abs(true_temp - pred_temp))

        # axs[j, 2].contourf(x, y, np.abs(true_temp - pred_temp), levels=10, cmap=error_cmap, vmin=0, vmax=vmax * 1.2)

        axs[j, 2].pcolormesh(x, y, np.abs(true_temp - pred_temp), antialiased=True, rasterized=True, cmap=error_cmap, vmin=0, vmax=vmax * 1.2)

        axs[j, 0].set_aspect(1)
        axs[j, 1].set_aspect(1)
        axs[j, 2].set_aspect(1)

        axs[j, 0].set_axis_off()
        axs[j, 1].set_axis_off()
        axs[j, 2].set_axis_off()


    axs[0, 0].set_title('Predicted')
    axs[0, 1].set_title('True')
    axs[0, 2].set_title('Error')

    # Write options to file name.
    file_name = f'heatmaps/multivariate_normal_seed_{seed}_n_steps_{n_steps}_n_true_mode_{n_true_mode}_n_error_mode_{n_error_mode}_cov_{cov_}_dx_{dx}_cmap_{ref_cmap}_error_cmap_{error_cmap_}.svg'

    plt.savefig(file_name)

    if plot:
        plt.show()

    plt.close('all')
```
